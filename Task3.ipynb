```python
#problem 1
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#problem 2
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
punctuation = ",.!?"
for p in punctuation:
    query = query.replace(p, "")

# Step 3: Split the sentence into words
words = query.split()

# Step 4: Count word frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Display keywords searched more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#problem 3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through readings with index
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#problem 4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Step 1: Count users per domain
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#problem 5
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect sales spikes
for day, value in enumerate(sales, start=1):
    if value > average_sales * 1.3:
        print(f"Day {day}: {value}")
```

```python
#problem6
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

# Step 1: Count occurrences of each user ID
for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Step 2: Display duplicate user IDs
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
